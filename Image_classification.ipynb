```python
import numpy as np
import pandas as pd
import tensorflow
from tensorflow import keras
```

```python
df=pd.read_excel("/content/dataset_final_neural.xlsx")
```

```python
df
```

```python
data=df.copy()
```

```python
data.info()
```

```python
data.dropna(axis=0,how="any",inplace=True)
```

```python
data.info()
```

```python
data.corr()
```

```python
data.drop("MSCN",axis=1,inplace=True)
```

```python
data.corr()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x=data.drop("Target",axis=1)
y=data["Target"]
```

```python
x1=pd.DataFrame(ss.fit_transform(x))
```

```python
x1
```

```python
pca=PCA(n_components=11)
pca.fit(x1)
pca.explained_variance_ratio_.cumsum()
```

```python
x1_pca=pd.DataFrame(pca.transform(x1),columns=["pca1","pca2","pca3","pca4","pca5","pca6","pca7","pca8","pca9","pca10","pca11"])
```

```python
x1_pca
```

```python
x1_pca.corr()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
xtr,xte,ytr,yte=train_test_split(x1_pca,y,test_size=0.2,random_state=1)
```

```python
lr=LinearRegression()
lr.fit(xtr,ytr)
print(lr.score(xtr,ytr))
print(lr.score(xte,yte))
```

```python
from sklearn.model_selection import GridSearchCV,KFold
fold=KFold(n_splits=5)
```

**using kfold and grid search**

```python
#on random forest algo
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(xtr,ytr)
print(model.score(xtr,ytr))
print(model.score(xte,yte))
```

```python
#got 97.76% on rf without parameter tuning
from sklearn.metrics import confusion_matrix,classification_report
model_rf=RandomForestClassifier(random_state=1,class_weight="balanced")
para={"max_depth":[5,10,15,20],
      "n_estimators":[25,50,75],
    "max_features":[3,5,7,9,10]
}
grid_rf=GridSearchCV(model_rf,param_grid=para,cv=fold,verbose=1)
grid_rf.fit(xtr,ytr)
print(grid_rf.best_estimator_)
y_pred=grid_rf.predict(xte)
print(classification_report(yte,y_pred))
print(grid_rf.score(xte,yte))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_gbt=GradientBoostingClassifier(random_state=1)
para_gb={"n_estimators":[50,75,100],
        "max_depth":[2,3,4],
        "learning_rate":[0.1,0.075,0.05]}
grid_gbt=GridSearchCV(model_gbt,param_grid=para_gb,cv=fold,verbose=1)
grid_gbt.fit(xtr,ytr)
print(grid_gbt.best_estimator_)
print(grid_gbt.score(xte,yte))
```

```python
#multi layer perceptron
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(12,input_shape=(11,),activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(6,activation="relu"))
model.add(Dense(2,activation="relu"))
model.add(Dense(1,activation="relu"))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xtr,ytr,epochs=150,batch_size=10)
```

```python
ypred1 = model.predict(xte)
print(ypred1)
ypred = (model.predict(xte) > 0.5).astype(int)
print(classification_report(yte,ypred))
```
